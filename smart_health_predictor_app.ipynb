importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
import joblib
```

Heart-disease

```python
heart_disease_data=pd.read_csv("Heart_disease.csv")
print(heart_disease_data.head())
print(heart_disease_data.info())
```

spliting independent and dependent data 

```python
X_heart_disease= heart_disease_data.drop("target", axis=1)
y_heart_disease= heart_disease_data["target"]
```

```python
print(X_heart_disease)
print(y_heart_disease)
```

train_test split

```python
x_train_heart,x_test_heart,y_train_heart,y_test_heart=train_test_split(X_heart_disease,y_heart_disease,test_size=0.2,random_state=42,stratify=y_heart_disease)
```

feature scaling

```python
# Scale ONLY the features
sc_heart = StandardScaler()
x_train_heart = sc_heart.fit_transform(x_train_heart)
x_test_heart = sc_heart.transform(x_test_heart)
```

training logistic Regression,xgboost,RandomForestClassifier

```python
# Ensure target is categorical (0 or 1)
y = heart_disease_data["target"].astype(int)
```

```python
print(y.unique())
print(y.value_counts())
```

```python
lr_heart=LogisticRegression()
lr_heart.fit(x_train_heart,y_train_heart)
y_pred_lr = lr_heart.predict(x_test_heart)
```

```python
# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train_heart, y_train_heart)
y_pred_rf = rf_model.predict(x_test_heart)

# XGBoost
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42)
xgb_model.fit(x_train_heart, y_train_heart)
y_pred_xgb = xgb_model.predict(x_test_heart)

# =======================
# 5. Evaluation
# =======================
print("Logistic Regression Accuracy:", accuracy_score(y_test_heart, y_pred_lr))
print("Random Forest Accuracy:", accuracy_score(y_test_heart, y_pred_rf))
print("XGBoost Accuracy:", accuracy_score(y_test_heart, y_pred_xgb))

print("\nClassification Report (Logistic Regression):\n", classification_report(y_test_heart, y_pred_lr))
print("\nClassification Report (Random Forest):\n", classification_report(y_test_heart, y_pred_rf))
print("\nClassification Report (XGBoost):\n", classification_report(y_test_heart, y_pred_xgb))

print("\nConfusion Matrix (Logistic Regression):\n", confusion_matrix(y_test_heart, y_pred_lr))
print("\nConfusion Matrix (Random Forest):\n", confusion_matrix(y_test_heart, y_pred_rf))
print("\nConfusion Matrix (XGBoost):\n", confusion_matrix(y_test_heart, y_pred_xgb))
```

saving the models

```python
import pickle
```

```python
with open("heart_model.pkl", "wb") as f:
    pickle.dump(xgb_model, f)
```

diabetes 

loading data 

```python
diabetes_dataset = pd.read_csv("datasets/diabetes.csv")
```

Replacing Zeros with NaN

```python
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
diabetes_dataset[cols_with_zero] = diabetes_dataset[cols_with_zero].replace(0, pd.NA)
```

 Imputing Missing Values

```python
diabetes_dataset.fillna(diabetes_dataset.median(), inplace=True)
```

marking dependent and independent data

```python
X_diabetes = diabetes_dataset.drop("Outcome", axis=1)
y_diabetes = diabetes_dataset["Outcome"]
```

train_test_split

```python
pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE

# Scale the features
X_scaled = sc_diabetes.fit_transform(X_diabetes)

# Apply SMOTE after scaling
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_scaled, y_diabetes)

# Train-test split on balanced data
X_train_diabetes, X_test_diabetes, y_train_diabetes, y_test_diabetes = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)
```

feature Scaling

```python
sc_diabetes = StandardScaler()
X_train_diabetes = sc_diabetes.fit_transform(X_train_diabetes)
X_test_diabetes = sc_diabetes.transform(X_test_diabetes)
```

Training Models

```python
import xgboost as xgb
# Define XGBoost model
xgb_model_diabetes = xgb.XGBClassifier(eval_metric="logloss", use_label_encoder=False)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
# Hyperparameter tuning
param_grid = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 4, 5],
    "learning_rate": [0.01, 0.05, 0.1],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0]
}

grid_search = GridSearchCV(
    estimator=xgb_model_diabetes,
    param_grid=param_grid,
    scoring="accuracy",
    cv=5,
    n_jobs=-1,
    verbose=1
)
```

```python
# Train with grid search
grid_search.fit(X_train_diabetes, y_train_diabetes)
```

```python
# Best model
best_xgb = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
```

evaluate

```python
# Predictions
y_pred = best_xgb.predict(X_test_diabetes)
```

```python

acc = accuracy_score(y_test_diabetes, y_pred)
print("✅ Improved Diabetes Model Accuracy:", acc)
print("\nClassification Report:\n", classification_report(y_test_diabetes, y_pred))
```

```python
import pickle

# Save model
with open("diabetes_model.pkl", "wb") as f:
    pickle.dump(best_xgb, f)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle
```

```python
kidney_dataset = pd.read_csv("datasets/kidney_disease(kaggle).csv")

# Display the first few rows of the dataset
kidney_dataset.head()
```

Drop id column if present

```python
if "id" in kidney_dataset.columns:
    kidney_dataset = kidney_dataset.drop("id", axis=1)
kidney_dataset.replace("?", np.nan, inplace=True)
```

Replace missing values

```python
kidney_dataset.replace("?", np.nan, inplace=True)

# Drop rows with missing values
kidney_dataset.dropna(inplace=True)
```

Separate features and target

```python
X_kidney = kidney_dataset.drop("classification", axis=1)
y_kidney = kidney_dataset["classification"]
```

Convert categorical columns to numeric using LabelEncoder

```python
for col in X_kidney.columns:
    if X_kidney[col].dtype == "object":
        le = LabelEncoder()
        X_kidney[col] = le.fit_transform(X_kidney[col])
```

```python
# Encode target
y_kidney = y_kidney.map({"ckd": 1, "notckd": 0})
```

Fill missing values with column mean

```python
kidney_dataset.fillna(kidney_dataset.mean(numeric_only=True), inplace=True)
```

Train-test split

```python
X_train_kidney, X_test_kidney, y_train_kidney, y_test_kidney = train_test_split(
    X_kidney, y_kidney, test_size=0.2, random_state=42
)
```

Scale features

```python
scaler = StandardScaler()
X_train_kidney = scaler.fit_transform(X_train_kidney)
X_test_kidney = scaler.transform(X_test_kidney)
rf_kidney = RandomForestClassifier(n_estimators=100, random_state=42)
```

Train model

```python
# Use the classification column directly (already 0/1)
y_kidney = kidney_dataset["classification"]

# Re-do train-test split to ensure y_train_kidney and y_test_kidney are correct
X_train_kidney, X_test_kidney, y_train_kidney, y_test_kidney = train_test_split(
	X_kidney, y_kidney, test_size=0.2, random_state=42
)

xgb_model_kidney = xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42)
xgb_model_kidney.fit(X_train_kidney, y_train_kidney)
```

```python
y_pred = xgb_model_kidney.predict(X_test_kidney)
print("✅ Kidney Model Accuracy:", accuracy_score(y_test_kidney, y_pred))
```
